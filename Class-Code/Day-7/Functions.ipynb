```python
greeting2('Muhammad Kamran Khan') # arguments
```

```python
starter = greeting3('Kamran')
```

```python
def sum(a, b): # a and b are parameters.
    c = a + b
    return c # returned value

# function calling
x = sum(10, 15) # 10 and 15 are arguments.
print(x)
```

```python
def sum_d(a, b):
    return a + b
```

```python
sum_d(14, 5)
```

```python
sum_d(14)
```

```python
def sum_d(a = 10, b = 20): # default parameters
    return a + b
```

```python
sum_d(4, 5)
```

```python
sum_d()
```

```python
def sum1(a, b):
    print('a:', a)
    print('b:', b)
    return a + b
```

```python
sum1(5, 6)
```

```python
sum1(b = 6, a = 5)
```
