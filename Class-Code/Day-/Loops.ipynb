### List Comprehension

```python
evens = []
for i in range(0, 20):
    if i % 2 == 0:
        evens.append(i)
    
```

```python
evens
```

```python
[i for i in range(0, 20) if i % 2 == 0]
```

```python
for x in ('apple', 'banana', 'mango', 'cherry'):
    print(f'Hello-{x}')
```

```python
fruits_dict = {'A': 'apple', 'B': 'banana', 
               'C': 'cherry', 'D': 'Dragon fruits'}
```

```python
for x in fruits_dict:
    print(x)
```

```python
for x in fruits_dict.values():
    print(x)
```

```python
for x in fruits_dict.items():
    print(x)
```

```python
a, b, c = 'apple', 'banana', 'cherry'
```

```python
print(a)
print(b)
print(c)
```

```python
for x, y in fruits_dict.items():
    print(f'{x}-{y}')
```

```python
# break
# continue
# pass
for i in range(0, 20):
    if i == 6:
        break
    print(i)
```

```python
for i in range(0, 20):
    if i == 6 or i == 7:
        continue
    print(i)
```

```python
for i in range(0, 20):
    pass
    print(i)
```

```python
for i in range(1, 5):
    for j in ['apple', 'banana', 'cherry']:
        print(f'{i}- {j}')
    print()
```
